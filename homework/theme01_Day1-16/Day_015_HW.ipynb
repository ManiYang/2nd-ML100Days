```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

# 忽略警告訊息
# %matplotlib inline
# import warnings
# warnings.filterwarnings('ignore')
```

```python
"""
Your Code Here
"""
matrix = np.random.random((10, 10)) * 2.0 - 1.0

plt.figure(figsize=(4, 3.5))
sns.heatmap(matrix, cmap='BrBG')
plt.show()
```

```python
"""
Your Code Here
"""
matrix = np.random.random((100, 3)) * 2.0 - 1.0
df = pd.DataFrame(matrix, columns=['A', 'B', 'C'])

# 隨機給予 0, 1, 2 三種標籤
df['label'] = np.random.choice(['0', '1', '2'], size=matrix.shape[0])
df.head()

# 繪製 seborn 進階 Heatmap
# grid = sns.PairGrid(data=df, size = 3, diag_sharey=False, 
#                     hue = 'index', vars = [x for x in list(plot_data.columns) if x != 'index'])
grid = sns.PairGrid(df, vars=['A', 'B', 'C'], hue='label', diag_sharey=False)
grid.map_upper(plt.scatter , alpha = 0.2)
grid.map_diag(plt.hist)
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)
grid = grid.add_legend()

plt.show()
```

```python
"""
Your Code Here
"""
matrix = np.random.randn(100, 3)
df = pd.DataFrame(matrix, columns=['A', 'B', 'C'])

# 隨機給予 0, 1, 2 三種標籤
df['label'] = np.random.choice(['0', '1', '2'], size=matrix.shape[0])
df.head()

# 繪製 seborn 進階 Heatmap
# grid = sns.PairGrid(data=df, size = 3, diag_sharey=False, 
#                     hue = 'index', vars = [x for x in list(plot_data.columns) if x != 'index'])
grid = sns.PairGrid(df, vars=['A', 'B', 'C'], hue='label', diag_sharey=False)
grid.map_upper(plt.scatter , alpha = 0.2)
grid.map_diag(plt.hist)
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)
grid = grid.add_legend()

plt.show()
```
